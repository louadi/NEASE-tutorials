```python
def exons_to_edges(mapped,G):
    mapped['domain']=mapped['NCBI gene ID'].astype(int).astype(str)+'/'+mapped['Pfam ID']
    mapped['Interacting_domain']=mapped['domain'].apply(lambda x: G.has_node(x) )
    return mapped
```

```python
def edge_enrich(a,b,p,n):
    #function to calculate P value
    #test if affected edges by AS are significally enriched in a pathway
    # fisher exact test
    # a+b affected adges, with a the one linked to pathway p
    #p total degree of pathway p
    #n total edges in the ppi used.
    
    #linked to pathway but not affected
    c=p-a
    
    # background of test: not linked to p and not affected edges
    d=(2*n)-p-b
    
    # retun oddsratio and pvalue from fisher exact test
    return   stats.fisher_exact([[a, b], [c, d]], alternative='greater')
```

```python
# General enrichment analysis

n=52467

sig=0
reject=0

log_pvalues={}
for path in paths:
    affected=0
    a=0
    b=0
    try:
        p=int(Pathways [Pathways['pathway']==path]['Degree in the structural PPI'])
        path_genes=list(Pathways [Pathways['pathway']==path]['entrez_gene_ids'])[0]
    except:
        pass
    for g in glist:
            affected=affected+len(glist[g])
            a=a+len([x for x in glist[g] if x in path_genes ])

    b=affected-a
        
    _,p_values=edge_enrich(a,b,p,n)

    if p_values<0.05:
            log_pvalues[path]=-np.log(p_values)
            sig=sig+1
            print(path)
            print(p_values) 
            print(a,b)

            print('\n')
    else: reject=reject+1
```

```python
# for enriched pathways
# count how many events is enriched in each one

n=52467

pathway_count={}


for path in list(log_pvalues):
    try:
        p=int(Pathways [Pathways['pathway']==path]['Degree in the structural PPI'])
        path_genes=list(Pathways [Pathways['pathway']==path]['entrez_gene_ids'])[0]
        pathway_count[path]=0
        for g in glist:

            affected=glist[g]
            a=len([x for x in affected if x in path_genes ])
            b=len(affected)-a

            _,p_values=edge_enrich(a,b,p,n)
            if p_values< 0.05:
                
                    pathway_count[path]+=1
                    
                    
                    print('The gene '+g2n[g]+ ' is  identified to be associated to '+path+':\n '+str(a+b)+
                          ' affected interactions with '+str(a)+' of them DCM related ( p-value='+str(p_values)+')\n'  )
    except:
                pass
```
